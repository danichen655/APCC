```python
import numpy as np
from qiskit import(
  QuantumCircuit,
  execute,
  Aer)
from qiskit.visualization import plot_histogram
simulator = Aer.get_backend('qasm_simulator')
circuit = QuantumCircuit(3, 3)
circuit.h(0)
circuit.cx([0],[1])
circuit.cx([1],[2])
circuit.measure([0], [0]) 
circuit.measure([1], [1])
circuit.measure([2], [2])
job = execute(circuit, simulator, shots=1000)
# Almacenamos los resultados
result = job.result()
# Capturamos las ocurrencias de salida
counts = result.get_counts(circuit)
# Escribimos el número de ocurrencias
print("\nNúmero de ocurrencias para 0 y 1:",counts)
# Dibujamos el circuito
circuit.draw()
```

El volumen cuántico es una métrica relacionada al rendimiento con las capacidades e índices de error de un computador cuántico, es decir, cuando mayor sea el volumen cuántico, el computador cuántico puede solucionar problemas más complejos y con más exactitud.
